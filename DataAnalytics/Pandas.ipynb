# working with large datasheet


when we working with very large data sheet then we divide data in chunk because some time our memory don't have do these computation and read

```python
for df in pd.read_csv("d.csv",chunksize=5):
    print("CHUNK DATA")
    print(df)
```

```python
new_df=pd.DataFrame(columns=df.columns)
for df in pd.read_csv("d.csv",chunksize=5):
    results=df.groupby(["year"]).count()
    new_df=pd.concat([new_df,results])
    
```
