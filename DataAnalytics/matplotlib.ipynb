```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#resize the graph
plt.figure(figsize=(5,3),dpi=300)#dpi =pix per inc
#use shorthand notation
#fmt='[color][marker][line]'
plt.plot(x,y,'r*--',label='2x')
#line two
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'g',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'y--',label='x^2')


# plt.plot(x,y,label='2x',linestyle='--',color='red',linewidth=2,marker='*',markersize=10,markeredgecolor='black')
plt.title("first graph",fontdict={"fontname":"Comic Sans MS"},fontsize=30)
plt.xlabel('x')
plt.ylabel('y')
#scale for graph
plt.xticks([0,1,2,3,4])
# plt.yticks([2,4,6,8,10])
plt.legend()#added level within graph
plt.savefig("first_graph.png",dpi=300)
plt.show()
```

# ## bar chart

```python
labels=['A','B','C']
values=[1,4,8]
bars=plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('o')
plt.savefig("bar.png",dpi=300)
plt.show()
```
