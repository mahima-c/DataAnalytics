
# TOKENIZATION

```python
## SENTENCE TOKENIZATION ##
# USING DIFFRENT NLTK INTERFACE

# First apporach
# using sents_tokenize function to perform tokenization
# this is default and recommended sentence tokenizer
# pre-trained in serval language models and work really well on other language 
# besides english
emma_sents=nltk.sent_tokenize(emma)
```

```python
len(emma_sents)
```

```python
# Note:
#     tokenizer does'nt just use period to delimit the sentences
#     its also consider punctuation and capitalization
emma_sents[0:5]
```

```python
# Second approach
# using default PunktSentenceTokenizer class to perfom tokenization
# sent_tokenize also instance of punktsentencetokenizer class
punkt_sen=nltk.tokenize.PunktSentenceTokenizer()
```

```python
# using tokenize method
emma_sents2=punkt_sen.tokenize(emma)
```

```python
emma_sents2[0:5]
```

```python
# Third approch
# using instance of Regexptokenize class
# specific regular expression -based pattern are used
pattern1=r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<![A-Z]\.)(?<=\.|\?|\!)\s'
```

```python
# the gaps parameter is true ,use to  find gaps between token
regex_st=nltk.RegexpTokenizer(pattern=pattern1,gaps=True)
```

```python
emma_sents3=regex_st.tokenize(emma)
```

```python
emma_sents3[0:5]
```

```python
## WORD TOKENIZATION ##
# Using nltk diffrent interface
# first apporch
# using word_tokenize function
# default and recommended tokenizer
emma_words=nltk.word_tokenize(emma)
```

```python
len(emma_words)
```

```python
emma_words[0:200]
```

```python
# second apporach:
# using TreeBankWordTokenizer class
# based on the penn treebank and uses various regular expression to tokenize the text
treebank_wt=nltk.TreebankWordTokenizer()

# one primary assumption :sentence tokenization perform already
# used to tokenize the sentence to word
emma_sents[0]
```

```python
emma_word2=treebank_wt.tokenize(emma_sents[0])
```

```python
emma_word2[0:500]#period apperes at the end of sentences
```

```python
# third apporach
# using Regexptokenizer class
# pettern to identify word
pattern2=r'\w+'
```

```python
regex_wt=nltk.RegexpTokenizer(pattern=pattern2,gaps=False)
```

```python
emma_word3=regex_wt.tokenize(emma)
```

```python
emma_word3[0:500]
```

```python
# pattern for identify gaps
pattern3=r'\s+'
```

```python
regex_wt1=nltk.RegexpTokenizer(pattern=pattern3,gaps=True)
```

```python
emma_word4=regex_wt1.tokenize(emma)
```

```python
emma_word4[0:500]
```

```python
# get start and end indices for each token
# using span_tokenize method
word_indice=list(regex_wt1.span_tokenize(emma))
```

```python
word_indice
```

```python
[emma[start:end] for start,end in word_indice]#similar output
```
